## 이미지의 입력

```matlab
img1 = imread('d:\lena.jpeg');
whos
```

## 이미지 보기

```matlab
figure(1), imshow(img1)
```

## 이미지 저장하기

```matlab
imwrite(img1, 'd:\lena1.png')
```

## 저장한 이미지 다시 불러오기

```matlab
img2 = imread('d:\lena1.png');
figure(2), imshow(img2)
```

## 이미지를 여러가지 파일로 저장하기

```matlab
clear all;
img1 = imread('d:\lena.jpeg');
for k1 = 1:10
imwrite(img1, ['d:\img', num2str(k1), '.jpg']);
end
```

## 여러 이미지 파일을 불러와서 출력하기

```matlab
clear all;
figure(1);
for k1=1:8
img1 = (k1/10)*im2double(imread(['d:\img', num2str(k1), '.jpg']));
subplot(2,4,k1);
imshow(img1);
title(['Image with k = ', num2str(k1/10)]);
end
```
