# 간단한 영상처리
## 밝기 변화시키기

```matlab
img1 = imread('d:\lena.jpeg');
whos
```

### imread로 영상을 읽으면 uint8 (부호가 없는 8비트 정수) 데이터형으로 저장된다.
### uint8의 데이터 범위는 0 ~ 255 사이의 정수, 0은 검은 색이고 255은 흰색이다.
### 영상처리를 위해서는 산술연산이 가능한 데이터형으로 변환해야 한다.
### im2double 명령은 uint8 데이터형을 double형으로 변환시켜준다.
### double의 데이터 범위는 0 ~ 1 사이의 실수, 0은 검은 색이고 1은 흰색이다.

```matlab
img1 = im2double(imread('d:\lena.jpeg'));
img2 = 2*img1;     % 밝기를 2배 증가시킴
figure(1);
subplot(1,2,1), imshow(img1), title('Original image')
subplot(1,2,2), imshow(img2), title('Bright image')
```

### 영상의 밝기를 2배 증가시키면 픽셀값이 1이 넘는 것들은 1로 제한되어 포화상태가 된다.
### 따라서 포화상태를 고려해서 밝기를 조절할 필요가 있다.
### 어두운 영상으로 만들 때는 나누기를 해 주면 된다.

```matlab
img3 = img1/2;
figure(1);
subplot(1,3,1), imshow(img1), title('Original image')
subplot(1,3,2), imshow(img2), title('Bright image')
subplot(1,3,3), imshow(img3), title('Dark image')
```

## 영상의 크기 변환

```matlab
clear all;
img1 = imread('d:\lena.jpeg');
whos
```

### 영상의 크기가 512x512이다.
### 영상의 크기를 가로, 세로 각각 2배로 증가시키면

```matlab
img2 = imresize(img1, 2);
figure(1)
subplot(1,2,1), imshow(img1), title('Original image')
subplot(1,2,2), imshow(img2), title('Enlarged image')
whos
```

### 영상의 크기를 가로, 세로 각각 1/2배로 감소시키면

```matlab
img3 = imresize(img1, 0.5);
subplot(1,2,1), imshow(img1), title('Original image')
subplot(1,2,2), imshow(img2), title('Small image')
whos
```

### 영상의 크기를 조절하는 imresize에서는 보간법으로 nearest, bilinear, bicubic(디폴트)가 있다.
### nearest의 화질이 가장 떨어지고, bicubic이 가장 좋다.
### 속도는 nearest가 가장 빠르고 bicubic이 가장 느리다.

```matlab
clear all;
img1 = imread('d:\lena.jpeg');
img2 = imresize(img1, 2, 'nearest');
img3 = imresize(img1, 2, 'bilinear');
img4 = imresize(img1, 2, 'bicubic');
figure(1)
subplot(2,2,1), imshow(img1), title('Original image')
subplot(2,2,2), imshow(img2), title('Nearest')
subplot(2,2,3), imshow(img3), title('Bilinear')
subplot(2,2,4), imshow(img4), title('Bicubic')
```

### 보다 자세히 화질을 비교하기 위해 100x100 픽셀만을 출력해 보면

```matlab
figure(2)
subplot(1,3,1), imshow(img2(301:400,501:600)), title('Nearest')
subplot(1,3,2), imshow(img3(301:400,501:600)), title('Bilinear')
subplot(1,3,3), imshow(img4(301:400,501:600)), title('Bicubic')
```

## 영상의 회전

```matlab
clear all;
img1 = imread('d:\lena.jpeg');
img2 = imrotate(img1, 45);     % 시계반대방향으로 45도 회전
figure(1)
subplot(1,2,1), imshow(img1), title('Original image')
subplot(1,2,2), imshow(img2), title('Rotated image')
```

### 영상의 회전에서는 보간법과 경계박스 옵션을 줄 수 있는데
### 보간법은 nearest(디폴트), bilinear, bicubic이 있다.
### 경계박스 옵션은 loose(디폴트), crop이 있다. 
### loose는 회전된 영상을 모두 볼 수 있게 영상의 크기를 크게 하는 것
### crop은 원영상의 크기와 동일하게 잘라버린다.

```matlab
img3 = imrotate(img1, 45, 'nearest');
img4 = imrotate(img1, 45, 'bilinear');
img5 = imrotate(img1, 45, 'bicubic');
img6 = imrotate(img1, 45, 'nearest', 'crop');
img7 = imrotate(img1, 45, 'bilinear', 'crop');
img8 = imrotate(img1, 45, 'bicubic', 'crop');
figure(2)
subplot(2,3,1), imshow(img3), title('Nearest and Loose')
subplot(2,3,2), imshow(img4), title('Bilinear and Loose')
subplot(2,3,3), imshow(img5), title('Bicubic and Loose')
subplot(2,3,4), imshow(img6), title('Nearest and Crop')
subplot(2,3,5), imshow(img7), title('Bilinear and Crop')
subplot(2,3,6), imshow(img8), title('Bicubic and Crop')
```

## 영상의 히스토그램

```matlab
clear all;
img1 = imread('d:\lena.jpeg');
[count, bin] = imhist(img1);     % count는 히스토그램의 도수, bin은 해당 픽셀 값 (0~255)
whos
figure, imhist(img1)
```

## 히스토그램 평활화 (Histogram Equalization)

```matlab
img2 = histeq(img1);
figure(2)
subplot(2,2,1), imshow(img1), title('Original image')
subplot(2,2,2), imshow(img2), title('Histogram Equalized image')
subplot(2,2,3), imhist(img1), title('Histogram of original image')
subplot(2,2,4), imhist(img2), title('Histogram of histogram equalized image')
```
